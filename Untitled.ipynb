```python
from scipy.io.wavfile import read
import matplotlib.pyplot as plt

# read audio samples
input_data = read("/home/user/Downloads/AO/OSR_us_000_0010_8k.wav")
audio = input_data[1]
plt.figure()
# plot the first 1024 samples
plt.plot(audio)
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Time")
# set the title  
plt.title("Sample Wav")
# display the plot
plt.show()
```
